## Python code for ATM (Automated Teller Machine)

### About Project
An automated teller machine (ATM) or cash machine is an electronic telecommunications device that enables customers of financial institutions to perform financial transactions, such as cash withdrawals, deposits, funds transfers, balance inquiries or account information inquiries, at any time and without the need for direct interaction with bank staff.

This ATM System project is written in Python. The project file contains a python script . This is a simple console based system which is very easy to use. This is a menu driven program. Talking about the system, pin for privacy is already asssigned and the balance is already given to the account . It contains a class having diferent functions which includes ● Check Balance , ● Deposit Fund , ● Withdraw Fund , ● Transfer Fund and ● Change Pin. Here, at first when the user runs the program , the system asks the service user wants ranging from 1 to 6. When users enters his/her desired input ,if the enterd service is in range then it will open that service panel. The system asks for the pin , when the pin matches with the system pin , it work and process the further program else it will show an error message and ask you whether you want to continue further more.
After each transaction it will print the transaction details along with updated balance.


### Flow of each Functions : 

##### 1.Check Balance


On entering 1, it takes us to the “Check Balance” option, where, after we’ve entered the correct PIN for our bank account, we are able to see how much money is in our bank account.

##### 2.Deposit Fund

On entering 2, we are taken to the "deposit fund" option of the ATM, where we’ll be able to deposit an amount into our bank account. Here, entereing some amount as the deposit amount, is immediately added to the bank balance.

##### 3. Withdraw Fund

On entering 3, it will load up the "withdraw fund"screen, wherein after we have entered the right PIN for our bank account, we will be able to enter an amount to withdraw from our account.

If , I have entered an amount greater than my available balance in account then the code tells me that I won’t be able to withdraw an amount because it is greater than my account balance and is giving me an error message. 

Else ,I am being prompted to enter the amount, it deducts the amount entered, and shows us the updated balance of money in our account.

##### 4.Transfer Fund

On pressing 5,the "Transfer Fund" section where it will ask for receiver's account number. after entering the account no. it will check whether the entered account no is valid by checking that it is of length 10. If it is valid then enter the amount to transfer and after that enter the pin.
It will check whether the pin matches ,if the pin matches and it will check that available balance is greater than the entered amount .if the amount is greater than the available balance then it will print an error else it will transfer the fund and will print the updated balance.

##### 5. Change Pin

Pressing 4 allows me to "Change Pin" where I can change my ATM PIN. Here, I am prompted to enter my old PIN, after which the program allows me to enter the new PIN I want and again rentering the pin to ensure that user is entering the exact pin he/she wants.
So after the PIN update, I will not be able to use my old PIN for any processes or transactions as long as my program is running this time. 


##### 6.  Exit

On pressing 6, you will be exited from atm system.

###### 2. Class and Class objects

A class is a user-defined blueprint or prototype from which objects are created. Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by their class) for modifying their state.

An Object is an instance of a Class. A class is like a blueprint while an instance is a copy of the class with actual values.
An object consists of : 

State: It is represented by the attributes of an object. It also reflects the properties of an object.

Behaviour: It is represented by the methods of an object. It also reflects the response of an object to other objects.

Identity: It gives a unique name to an object and enables one object to interact with other objects.

######  __init__

The __init__ method is similar to constructors in Pyhton. Constructors are used to initialize the object’s state. Like methods, a constructor also contains a collection of statements(i.e. instructions) that are executed at the time of Object creation. It runs as soon as an object of a class is instantiated. The method is useful to do any initialization you want to do with your object.



```python
class Atm:                                    #using Atm as the class name .

    def __init__(self):                       # constructor
        self.pin = 1234                       # predefining the pin for atm .
        self.balance = 50000                  # predefining the balance for saving account .
        self.menu()                           # calling menu function
        
        
        

    def Check_balance(self):                  # this function checks the available balance in the account.
        input_pin = (input("Enter your pin: "))         # user enters the pin
        if int(input_pin) == self.pin:                          # condition check
            print("-"*70)                                  # this is just for seprating the output with other lines
            print("** Authetication success **")
            print(
                f"Your saving account balance is {self.balance} at {datetime.datetime.now()}")        # output
                            #datetime function will print timestamp of this  event.
                            #Timestamp is the date and time of occurrence of an event.
            print("-"*70)
        else:                                              # if condition fails
            print("-"*70)
            print("** Wrong Pin Enterd ** \n   --> Enter valid 4 digit pin ")                         # output
            print("-"*70)

        self.repeat()                                     # calling repeat function
        
        
        

    def Deposit_fund(self):                 # this function deposits the fund in the account.
        input_pin = (input("Enter your pin: "))           # input
        if int(input_pin) == self.pin:
            input_amount = int(input("Enter the amount to Deposit: "))
            self.balance += input_amount  # when user deposits fund then balance increase by the depositing amount
            print("-"*70)
            print("** Deposit Successful **")                # output message
            time.sleep(1)                # generating a delay of given time(in seconds) before going to next statement.
            print(
                f"Deposited Amount - {input_amount} at {datetime.datetime.now()}")                   # output
            time.sleep(0.4)
            print(
                f"Available Balance - {self.balance} at {datetime.datetime.now()}")                  # output
            print("-"*70)

        else:
            print("-"*70)
            print("** Wrong Pin Enterd ** \n   --> Enter valid 4 digit pin ")                        # output
            print("-"*70)

        self.repeat()                       # calling repeat function
        
        
        

    def Withdraw_fund(self):                # this function withdraws fund from the account
        input_pin = (input("Enter your pin: "))                            # input
        if int(input_pin) == self.pin:
            input_amount = int(input("Enter the amount to Withdraw: "))       #input
            if input_amount <= self.balance:                # condition check
                print("-"*70)                               # this is just for seprating the output with other lines
                print("** Withdrawl Successful **")                                                 # output
                self.balance -= input_amount
                time.sleep(1)
                print(
                    f"Withdrawl Amount - {input_amount} at {datetime.datetime.now()}")              # output
                time.sleep(0.4)
                print(
                    f"Available Balance - {self.balance} at {datetime.datetime.now()}")             # ouput
                print("-"*70)
            else:
                time.sleep(0.4)                            # this is just for seprating the output with other lines
                print("-"*70)
                print("** Withdrawl Unsuccessful ** \n   --> Insufficient Balance ")                # output
                print("-"*70)
        else:
            print("-"*70)
            print("** Wrong Pin Enterd ** \n   --> Enter valid 4 digit pin ")                       # output
            print("-"*70)

        self.repeat()                     # calling repeat function
        
        
        

    def Transfer_Fund(self):              # this function transfers fund from user account to other account
        input_account = str(int(input("Enter Receiver's 10 digit Account Number: ")))              # input
        if len(input_account) == 10:        # condition that account no. should be of exactly 10 digits
            input_amount = int(input("Enter the amount to Transfer: "))                            # input
            input_pin = (input("Enter your pin: "))
            if int(input_pin) == self.pin:        
                if input_amount <= self.balance:    # available balance should be graeter than input amount
                    print("-"*70)
                    print("** Transfer Successful **")                                             # output
                    self.balance -= input_amount    # available balance will decrease by transferred amount
                    time.sleep(1)
                    print(
                        f"Transferred to Account - {input_account} at {datetime.datetime.now()}")  # output
                    print(
                        f"Transferred Amount - {input_amount} at {datetime.datetime.now()}")       # output
                    time.sleep(0.4)
                    print(
                        f"Available Balance - {self.balance} at {datetime.datetime.now()}")        # output
                    print("-"*70)
                else:
                    print("-"*70)
                    print("** Withdrawl Unsuccessful ** \n   --> Insufficient Balance ")           # output
                    print("-"*70)
            else:
                print("-"*70)
                print("** Wrong Pin Enterd ** \n   --> Enter valid 4 digit pin ")                  # output
                print("-"*70)
        else:
            print("-"*70)
            print('''** Invalid Receiver's Account Number **\n   
            --> Enter valid 10 digit Receiver's Account Number''')  # output
            print("-"*70)
        
        self.repeat()               # calling repeat function
        
        
        

    def Change_Pin(self):           # function to change pin of the atm
        old_pin = (input("Enter previous Pin: "))                 #input    
        if int(old_pin) == self.pin:                                      # condition to match the pin
            new_pin = (input("Enter New Pin: "))
            new_pin1 = (input("Re-Enter New Pin: ")) # taking pin two times to 
            if int(new_pin) == int(new_pin1) and len(str(new_pin)) == 4:    # the length of pin must be of 4 digit
                self.pin = new_pin                              # updating pin with new pin
                print("-"*70)
                print(
                    f"Pin updated Successfully at {datetime.datetime.now()}")                  # output
                print("-"*70)
            else:
                print("-"*70)
                print("** New Pin doesn't match **\n   --> Enter similar 4 digit Pin ")          # output
                print("-"*70)
        else:
            print("-"*70) 
            print("** Wrong Pin Enterd ** \n   --> Enter valid 4 digit pin ")                    # output
            print("-"*70)
        
        self.repeat()            # calling repeat function


    def repeat(self):            # repeat functions asks whether the user wants to carry on using the services .
        time.sleep(1.2)

        print("Do you want to continue ? \n--> Enter 'Y' for yes or 'N' for no . ")# displaying message for user
        choice = (input("Enter your choice: "))     # input
        if choice == "Y" or choice == "y":            # condition check
            self.menu()                           # if y/Y then menu function is called
        elif choice == "N" or choice == "n":          # if n/N then it will return 
            print("-"*70)
            print("Exiting...")
            time.sleep(1)
            print("Thnak you for choosing DSAT Bank of PSIT")
            print("== Have a' good day !! ==")                                                         # output
            print("-"*70)
            return                                # to terminate the program
        else:
            print("-"*70)
            print("** Invalid choice **")                                                       # output
            print("-"*70)

            self.repeat()         # calling repeat function


    def menu(self):            # this is menu function. this will show all the services offerd by the atm system.
        print("""
            == Welcome to DSAT Bank of PSIT !! ==
                 1. Check Balance
                 2. Deposit Fund
                 3. Withdraw Fund
                 4. Transfer Fund
                 5. Change Pin
                 6. Exit
            """)                 # this is the various services 

        option = (input("Enter your choice: "))           # user inputs the choice of service

        if int(option) == 1:                                          # if user enters 1
            self.Check_balance()                                  # check_balance function is called

        elif int(option) == 2:                                        # if user enters 2
            self.Deposit_fund()                                  # deposit_fund function is called

        elif int(option) == 3:                                        # if user enters 3
            self.Withdraw_fund()                                  # withdraw_fund function is called

        elif int(option) == 4:                                        # if user enters 4
            self.Transfer_Fund()                                  # transfer_fund function is called

        elif int(option) == 5:                                        # if user enters 5
            self.Change_Pin()                                     # change_pin function is called

        elif int(option) == 6:                                        # if user enters 6
            print("-"*70)
            print("Exiting...")                                    # output message
            time.sleep(1)        # generating a delay of given time(in seconds) before going to next statement.
            print("Thnak you for choosing DSAT Bank of PSIT")
            print("== Have a good day !! ==")
            print("-"*70)
            return                                                 # to terminate the program
        else:                                                  # if user enters invalid option
            print("-"*70)
            print("** Invalid Choice ** \n   --> Restart System Again  ")     # output message
            print("-"*70)
            return                                                 # to terminate the program


bank=Atm()


```
